```python
import pandas as pd

# Simulação inicial de respostas
respostas = pd.DataFrame({
    "questao": ["Q1", "Q2", "Q3", "Q4", "Q5"],
    "dificuldade": ["facil", "facil", "media", "dificil", "dificil"],
    "resposta": [1, 0, 1, 1, 0],
    "peso": [1, 1, 2, 3, 3]
})

# Pontuação base
respostas["pontuacao"] = respostas["resposta"] * respostas["peso"]

pontuacao_total = respostas["pontuacao"].sum()

print("Pontuação total:", pontuacao_total)
respostas
```
